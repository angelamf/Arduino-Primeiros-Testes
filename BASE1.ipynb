```
# Valores Nulos
def testa_valores_nulos(dataframe, atributo): 
    if sum(dataframe[atributo].isnull()) == 0:
        print(atributo," => OK (não tem valores nulos)")
    else:
        print(atributo," => *** TEM VALORES NULOS")

for i in range(0, len(atributos)):
    testa_valores_nulos(df, atributos[i])
```

```
# Valores '(null)'
def testa_valores_null(dataframe, atributo): 
    if sum(dataframe[atributo] == '(null)') == 0:
        print(atributo," => OK (não tem valores '(null)')")
    else:
        print(atributo," => *** TEM VALORES '(null)'")

for i in range(0, len(atributos)):
    testa_valores_null(df, atributos[i])
```

```
#br  => *** TEM VALORES NULOS
#km  => *** TEM VALORES NULOS

# Eliminando amostras que tem br e km nulos
df = df[df['br'].notna()]
df = df[df['km'].notna()] # Nem precisava, onde br é NaN, km também é

testa_valores_nulos(df, 'br')
testa_valores_nulos(df, 'km')
```

```
# Testa se os tipos são iguais

def testa_tipos_str(dataframe, atributo):    
    if sum(dataframe[atributo].map(type) != str) == 0: 
        print(atributo," => todos são str")
    else: 
        print(atributo," => *** ALGUNS NÃO SÃO STR")

def testa_tipos_int(dataframe, atributo):    
    if sum(dataframe[atributo].map(type) != int) == 0: 
        print(atributo," => todos são int")
    else: 
        print(atributo," => *** ALGUNS NÃO SÃO INT")

def testa_tipos_float(dataframe, atributo):    
    if sum(dataframe[atributo].map(type) != float) == 0: 
        print(atributo," => todos são float")
    else: 
        print(atributo," => *** ALGUNS NÃO SÃO float")

######################################
print("Atributos que devem ser STR")
for i in range(0, len(atributos_str)):
    testa_tipos_str(df, atributos_str[i])

print("*"*50)
print("Atributos que devem ser INT")
for i in range(0, len(atributos_int)):
    testa_tipos_int(df, atributos_int[i])

print("*"*50)

print("Atributos que devem ser FLOAT")
for i in range(0, len(atributos_float)):
    testa_tipos_float(df, atributos_float[i])
```

```
# Testanto se tem aquele tipo naquele atributo
def testa_se_tem_tipos_str(dataframe, atributo):    
    if sum(dataframe[atributo].map(type) == str) > 0: 
        print(atributo," => tem algum str")
        print("*"*25)

def testa_se_tem_tipos_int(dataframe, atributo):    
    if sum(dataframe[atributo].map(type) == int) > 0: 
        print(atributo," => tem algum int")
        print("*"*25)

def testa_se_tem_tipos_float(dataframe, atributo):    
    if sum(dataframe[atributo].map(type) == float) > 0: 
        print(atributo," => tem algum float")
        print("*"*25)

######################################
print("Atributos que devem ser INT")
for i in range(0, len(atributos_int)):
    testa_se_tem_tipos_str(df, atributos_int[i])
    testa_se_tem_tipos_float(df, atributos_int[i])

print("*"*50)
print("Atributos que devem ser FLOAT")
for i in range(0, len(atributos_float)):
    testa_se_tem_tipos_str(df, atributos_float[i])
    testa_se_tem_tipos_int(df, atributos_float[i])
```

Conclusão: Esses atributos acima, tem que ser convertidos


```
# converte float em int
def converte_float_to_int(dataframe, atrib):
    for i in range(0, len(atrib)):
        dataframe[atrib[i]] = dataframe[atrib[i]].astype(int)

converte_float_to_int(df, atributos_int)

# converte srt em float
def converte_srt_to_float(dataframe, atrib):
    for i in range(0, len(atrib)):
        dataframe[atrib[i]] = dataframe[atrib[i]].apply(lambda x: x.replace(",", "."))
        dataframe[atrib[i]] = dataframe[atrib[i]].astype(float)

converte_srt_to_float(df, atributos_float)
```

```
# 4. Qual estado brasileiro com maior índice de acidentes?
def estado_maior_indice_acidente(dataframe, ano):
  return  dataframe[dataframe['data_inversa'].str.contains(ano)].groupby('uf').count().id.sort_values(ascending=False)

print('Ano de 2017\n', estado_maior_indice_acidente(df, '2017'))
print('Ano de 2018\n', estado_maior_indice_acidente(df, '2018'))
print('Ano de 2019\n', estado_maior_indice_acidente(df, '2019'))
print('Ano de 2020\n', estado_maior_indice_acidente(df, '2020'))
```

**Conclusão**: Durante os 4 anos (2017, 2018, 2019 e 2020), Minas Gerais foi o estado com o maior número de acidentes.

```
# 7. Sob qual condição climática é mais seguro viajar?
total_acidentes = len(df['id'])

def condicao_climatica_mais_segura(dataframe):
  return dataframe.groupby('condicao_metereologica').count().drop('Ignorado', axis=0).id.sort_values(ascending=True)

condicao_climatica_mais_segura(df)
```

```
# 2. A maior média de acidentes ocorre aos finais de semana ou durante a semana?(Por ano)

def media_acidentes_semana_fds(dataframe, ano):
  return dataframe[dataframe['data_inversa'].str.contains(ano)].groupby('dia_semana').mean().id.sort_values(ascending=False)

print('Ano de 2017\n', media_acidentes_semana_fds(df, '2017'), '\n')
print('Ano de 2018\n', media_acidentes_semana_fds(df, '2018'), '\n')
print('Ano de 2019\n', media_acidentes_semana_fds(df, '2019'), '\n')
print('Ano de 2020\n', media_acidentes_semana_fds(df, '2020'))
```

**Conclusão:** 
- No ano de 2017 a maior média de acidentes ocorreu aos finais de semana. (Sábado)
- No ano de 2018 a maior média de acidentes ocorreu aos finais de semana. (Domingo)
- No ano de 2019 a maior média de acidentes ocorreu aos finais de semana. (Domingo)
- No ano de 2020 a maior média de acidentes ocorreu na semana. (Sexta)

```
# 8 - Qual a principal causa, tipo, classificacao_acidente de acidente no trânsito?(Por ano)
ANO = '2017'

def principal_causa_acidente(dataframe, ano):
  return dataframe[dataframe['data_inversa'].str.contains(ano)].groupby('causa_acidente').count().id.sort_values(ascending=False)

def principal_tipo_acidente(dataframe, ano):
  return dataframe[dataframe['data_inversa'].str.contains(ano)].groupby('tipo_acidente').count().id.sort_values(ascending=False)

def principal_classificacao_acidente(dataframe, ano):
  return dataframe[dataframe['data_inversa'].str.contains(ano)].groupby('classificacao_acidente').count().id.sort_values(ascending=False)

print('Principal causa de acidente em {}\n'.format(ANO), principal_causa_acidente(df, ANO), '\n')
print('Principal tipo de acidente em {}\n'.format(ANO), principal_tipo_acidente(df, ANO), '\n')
print('Principal classificação de acidente em {}\n'.format(ANO), principal_classificacao_acidente(df, ANO))
```
